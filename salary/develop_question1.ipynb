```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
from scipy import stats
from sklearn.datasets import load_iris
import pandas as pd

project_data_to_load = "salary_data_states1.xlsx"

project_data = pd.read_excel(project_data_to_load)

```

```python
project_data
```

```python
fig, ax = plt.subplots()
ax.scatter(project_data['JOB_TITLE_SUBGROUP'], project_data['PAID_WAGE_PER_YEAR'])
ax.set_xlabel('Job Title')
ax.set_ylabel('Yearly Wage')
ax.set_title('Job vs. Pay')
ax.tick_params(axis='x', rotation=45)
```

```python
job_avg_prevailing_wage = project_data.groupby('JOB_TITLE_SUBGROUP')['PREVAILING_WAGE_SUBMITTED'].mean()
job_avg_prevailing_wage

#group by job title to categorize job groups. looked at 2nd column and calculated the average for the offered wages for each sub group 
```

```python
cp_avg_prevailing_wage = project_data.groupby('EMPLOYER_NAME')['PREVAILING_WAGE_SUBMITTED'].mean()
cp_avg_prevailing_wage

#same as above but instead job title :D 
```
